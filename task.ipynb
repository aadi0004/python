```python
class User:
    def __init__(self, user_id, name, preferences, budget):
        self.user_id = user_id
        self.name = name
        self.preferences = preferences
        self.budget = budget

    def update_budget(self, amount):
        self.budget += amount


class Product:
    def __init__(self, product_id, name, category, price):
        self.product_id = product_id
        self.name = name
        self.category = category
        self.price = price


class ShoppingAssistant:
    def __init__(self):
        self.products = []  # List of all available products
        self.cart = []      # User's shopping cart

    def add_product(self, product):
        self.products.append(product)



    def recommend_products(self, user):
      return [
        product for product in self.products
        if product.category in user.preferences and product.price <= user.budget
    ]


    def add_to_cart(self, user, product_id):
        for product in self.products:
            if product.product_id == product_id and product.price <= user.budget:
                self.cart.append(product)
                user.update_budget(-product.price)
                print(f"{product.name} Added to cart.")
                return
        print("Product not found or no budget.")

    def view_cart(self):
        print("Shopping Cart:")
        for product in self.cart:
            print(f"{product.name} Price: ${product.price}")
        print(f"Total: ${sum(product.price for product in self.cart)}")

    def provide_insights(self, user):
        total_spent = sum(product.price for product in self.cart)
        savings = user.budget
        print(f"Total Spent: ${total_spent}")
        print(f"Remaining Budget: ${savings}")

```

```python

user = User(user_id=1, name="Alice", preferences=["electronics", "clothing"], budget=1200)

assistant = ShoppingAssistant()

# Add Products
assistant.add_product(Product(1, "Smartphone", "electronics", 300))
assistant.add_product(Product(2, "T-shirt", "clothing", 20))
assistant.add_product(Product(3, "Laptop", "electronics", 800))
assistant.add_product(Product(4, "Jeans", "clothing", 50))

print("Recommended Products:")
for product in assistant.recommend_products(user):
    print(f"{product.name} - ${product.price}")

assistant.add_to_cart(user, 1)  # Adding Smartphone
assistant.add_to_cart(user, 2)  # Adding T-shirt

assistant.view_cart()

assistant.provide_insights(user)
```
