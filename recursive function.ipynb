```python
class factorial:
    def calculate(self,n):
        # base case : factorial of 0 and 1 is 1
        if n==0 or n==1:
            return 1
        else:
            # recurrsive case n* factorial of n-1
            return n*self.calculate(n-1)
        
cal=factorial()
print("Factorial is: ",cal.calculate(5))
```

```python
def number_gen():
    for i in range(1,6):
        yield i
        
# using generator
gen = number_gen()
for num in gen :
    print(num)
```

```python
def number_gen(n):
    for i in range(1,n+1):
        yield i * i
        
# using generator
gen = number_gen(5)
for squ in gen :
    print(squ)
```

```python
# Class: Student
class Student:
    def __init__(self, name, preferred_topics):
        self.name = name
        self.preferred_topics = preferred_topics
        self.progress = {}

    def enroll_course(self, course):
        self.progress[course.title] = 0

    def update_progress(self, course, progress):
        self.progress[course.title] = min(progress, 100)

    def get_report(self):
        print(f"\nReport for {self.name}:")
        for course, progress in self.progress.items():
            print(f"{course}: {progress}%")


# Class: Course
class Course:
    def __init__(self, title, topic):
        self.title = title
        self.topic = topic


# Recommend Courses
def recommend_courses(student, courses):
    print(f"\nRecommended Courses for {student.name}:")
    for course in courses:
        if course.topic in student.preferred_topics:
            print(f"- {course.title} ({course.topic})")


# Main
python_basics = Course("Python Basics", "Python")
data_science = Course("Data Science", "Data Science")
ml_course = Course("Machine Learning", "AI")

courses = [python_basics, data_science, ml_course]
student = Student("Alice", ["AI", "Data Science"])

student.enroll_course(python_basics)
student.update_progress(python_basics, 50)
recommend_courses(student, courses)
student.get_report()
```
