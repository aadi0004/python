```python
# # Case Study 2: online Examination System

# problem: create a system for online examinations where :

# student scores are private and can not be accessed directly.

# Scores can only be added or retrived using specific methods.

# implement a feature to calculate the average score  of the student
```

```python
class online_exam:
  def __init__(self,name):
    self.name=name
    self.__score=[78,37,69,90]

  def show_score(self):
    return self.__score

  def add_score(self,num):
    if(num>0 and num<=100):
      self.__score.append(num)
      return f"{num} added"
    else:
      return "can not add number"

  def average(self):
    sum1=sum(self.__score)
    length=len(self.__score)
    return f"Average is: {sum1/length}"
```

```python
a=online_exam('asd')
print(a.name)
print(a.show_score())
print(a.add_score(23))
print(a.average())
```

```python
class stu():
  location= 'jaipur'

  def __init__(self,name,age):
    self.name=name
    self.age=age

  def getinfo(self):
    print(f"name of student is {self.name} and his age: {self.age}")

  def status(self):
    print("ok ok")


class traning(stu):
  def __init__(self,batch,duration):
    self.batch=batch
    self.duration=duration

  def getinformation(self):
    print(f"batch is {self.batch} and duration is {self.duration} hours")

a=stu("rahul",23)
a.getinfo()
a.status()

b=traning("DS",10)
b.getinformation()
b.status()  # inherite
```

```python
# multiple inheritance
class engine:
  def start(self):
    print("engine started")

class wheels:
  def rerote_wheel(self):
    print("wheels are rotating")

class car(engine,wheels):
  def drive(self):
    print("car is driving")

car=car()
car.start()
car.rerote_wheel()
car.drive()
```

```python
# multiple inheritance
class animal:
  def eat(self):
    print("animal is eating")

class dog(animal):
  def bark(self):
    print("dog is barking")

class cat(dog):
  def meow(self):
    print("cat is meowing")


cat=cat()
cat.eat()
cat.bark()
cat.meow()
```
